Fazendo [cadastro na plataforma](https://www.kaggle.com/account/login?phase=startRegisterTab&returnUrl=%2F) e clicando no link da competição, podemos ver uma descrição dos dados e das variáveis.

Como são muitas colunas, podemos tentar tirar uma primeira impressão sobre a informação de forma mais resumida.

````python
# listando colunas com valores nulos
dataset_com_nulos.isnull().sum()[dataset_com_nulos.isnull().sum() > 0]
````

````python
# removendo valores nulos de forma genérica
dataset_sem_nulos = dataset_com_nulos.dropna() # joga fora todas as linas com pelo menos um valor nulo
print(f"tamanho do dataset com nulos: {dataset_com_nulos.shape}")
print(f"tamanho do dataset sem nulos: {dataset_sem_nulos.shape}")

dataset_sem_nulos.isnull().sum()[dataset_sem_nulos.isnull().sum() > 0] # série vazia = nenhuma coluna com nulos
# o problema é que nesse caso não sobrou linha nenhuma (0 (linhas), 81 (colunas))
````

````python
# preenchendo valores nulos com alguma métrica da propria coluna
# uma estrategia mais interessante que simplesmente jogar a linha fora por ter um nulo, 
# poderia ser, por exemplo, preencher o nulo com o valor que mais se repete naquela coluna, a moda
novo_dataset_sem_nulos = dataset_com_nulos.copy() # iniciando o dataframe a ser preeenchido
for column in dataset_sem_nulos.columns:
    novo_dataset_sem_nulos[column] = novo_dataset_sem_nulos[column].fillna(dataset_com_nulos[column].mode()[0])

print(f"tamanho do dataset sem nulos: {novo_dataset_sem_nulos.shape} \n")
novo_dataset_sem_nulos.isnull().sum()[novo_dataset_sem_nulos.isnull().sum() > 0] 

# agora sim um dataset sem nulos, mas com o tamanho do original
display(novo_dataset_sem_nulos)
````

Do mesmo modo que lá em cima usamos as list comprehensions, aqui temos uma outra novidade: as funções lambda ou funções anônimas.

Basicamente essa é uma sintaxe que resume uma função de um argumento em uma linha:

```python
def x_na_quinta_padrao(x):
    return x ** 5
```

pode ser resumida como:

```python
lambda x: x ** 5
```

quando dentro de um contexto como no caso do apply que passa os valores dinamicamente para dentro do argumento x ou de forma bem menos usual como:

```python
lambda_x_na_quinta = lambda x: x ** 5
```

que pode ser acessada em qualquer lugar do cósigo, mas perde a sua vantagem que é ser definida e invocada em um só comando.

Nesse caso, `x_na_quinta_padrao` e `lambda_x_na_quinta` têm exatamente o mesmo comportamento.

Para entender um pouco sobre essa forma de escrever funções de uma linha, você pode dar uma lida [nesse post](https://giovannireisnunes.wordpress.com/2018/08/03/funcoes-anonimas-em-python/) ou [nesse vídeo](https://www.youtube.com/watch?v=EQRsJ2bKvUU) do YouTube.

````python
boston_housing_df["OverallQual_bubble"] = boston_housing_df['OverallQual'].apply(lambda x: x**5)
````

## Regressão

Vamos seguir usando o dataset boston housing, mas para simplificar um pouco a análise, vamos selecionar apenas 2 colunas: LotArea e OverallQual para prever SalePrice.

Usaremos o modelo de [regressão linear do Sciki-learn](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

````python
boston_housing_reg = boston_housing_df[['LotArea', 'OverallQual', 'SalePrice']]
boston_housing_reg
````

### Linear

````python
from sklearn.linear_model import LinearRegression

# separando as variáveis de entrada das variável a ser prevista
x = boston_housing_reg[['LotArea', 'OverallQual']]
y = boston_housing_reg['SalePrice']

reg = LinearRegression() # instanciando o modelo
reg.fit(x, y) # treinando o modelo

prev_indice_1458 = reg.predict([[9717,5]])[0] # fazendo uma previsão da casa de indice 1458 do dataframe
print(prev_indice_1458) # selecionamos o indice 0 duas vezes por causa do formato de saída dos dados
````

Mas porque o modelo não acerta uma casa que ele conhece?\
Lembrando dos vídeos, uma regressão é um modelo que ajusta uma curva aos dados tentando minimizar o erro então provavelmente não vai passar exatamente em cima de algum dos pontos usados para o treino.

````python
boston_housing_reg.loc[1458]
````

### Medidas de desempenho da regressão

Para avaliar o modelo de uma forma um pouco mais cuidadosa que apensa verificar se o indice 1458 está suficientemente próximo do valor real, vamos retreinar o modelo com um conjunto de dados e testar como esse modelo se comporta lidando com dados que ele nunca viu:

````python
from sklearn.model_selection import train_test_split

x = boston_housing_reg[['LotArea', 'OverallQual']]
y = boston_housing_reg['SalePrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7) 

print(x_train.shape) # duas colunas com 70% dos dados
print(y_train.shape) # uma coluna com 70% dos dados
print(x_test.shape)
print(y_test.shape)
````

````python
reg = LinearRegression() # instanciando o modelo
reg.fit(x_train, y_train) # treinando o modelo

y_pred = reg.predict(x_test) # prevendo todos os valores de teste de uma vez

results = x_test.copy()
results['y_real'] = y_test
results['y_pred'] = y_pred

display(results.head(20)) # comparando o resultado de 20 previsoes
````

Podemos querer dar uma olhada nos erros pra tentar facilitar as contas

````python
results['erro_absoluto'] = results['y_pred'] - results['y_real']
results['erro_percentual'] = (results['y_pred'] - results['y_real'])/results['y_pred']
results
````

Pode ser que dependendo da forma com que os dados foram aleatorizados e quebrados em treino e teste nós tenhamos valores muito discrepantes nos extremos do nosso dataset de erros. Pra fugir um pouco disso nesse notebook, vamos eliminar os 1% maiores es 1% menores erros.

````python
results_limpos = results[(results['erro_percentual'] > 0.01) & (results['erro_percentual'] < 0.99)]
results_limpos
````

#### Intervalo de confiança do erro médio

````python
results_limpos.sort_values(by=['erro_percentual'])
````

````python
print(results_limpos['erro_percentual'].describe())
````

````python
plt.figure(figsize=(15,10)) # tamanho do gráfico
sns.distplot(results_limpos['erro_percentual']) # dados a serem apresentados

plt.xlabel("Erro", fontdict={"size": 15})
plt.title("Distribuição dos erros percentuais", fontdict={'size': 20})
 
plt.show() 
````

Em um modelo de regressão bem comportado, queremos que nosso erro médio seja 0, mas muito mais importante que um erro médio é a viariância dessa média também ser o menor possível.

#### Precisão

Precisão é a razão $ tp/(tp + fp) $ onde tp são os verdadeiros positivos e fp são os falsos positivos. É a capacidade que o modelo tem de acertar quando ele prevê que algo é positivo. O melhor valor possível é 1 e o pior 0.


````python
from sklearn.metrics import precision_score

csl_precision_score = precision_score(results['y_real'], results['y_pred'])
csl_precision_score
````

#### Revocação (recall)

Revocação é a razão $tp / (tp + fn)$ onde tp são os verdadeiros positivos e fn, os falsos negativos. Essa métrica avalia a capacidade do modele acertar o que de fato é verdadeiro.

O melhor valor possível é 1 e o pior 0.

````python
from sklearn.metrics import recall_score

cls_recall_score = recall_score(results['y_real'], results['y_pred'])
cls_recall_score
````

Temos aqui um modelo que acerta muito pela precisão, acerta um pouco menos das vezes que diz que uma cassa é boa e identifica com sucesso menos ainda das casas que são de fato boas.

Porque temos tantas métricas diferentes pra um mesmo problema de classificação?

A resposta, não muito intuitiva, é **dados desbalanceados**.

Quando fizemos a separação entre o que seria uma casa considerada boa e uma casa considerada ruim nesse nosso exemplo simplista, dissemos que apenas casas com OverallQual acima de 7 seriam consideradas boas. Vamos dar uma olhadinha em como ficou a nossa distribuição de BinnaryOverallQual?

````python
results['y_real'].mean()
````

Parece que situação é um pouquinho pior que o que a gente esperava né? Se as distribuições das classes fosse uniforme, 30% seriam consideradas "boas" mas apenas algo em torno da metade na verdade caiu nesse grupo e agora as diferenças nas métricas faz sentido.

#### F1

Para tentar contemplar esse desbalanço entre classes, uma das abordagens é o f1 score, que é uma média harmônica entre precisão e revocação: $ 2*(precisao * revocação) / (precisão + revocação) $.

````python
from sklearn.metrics import f1_score

cls_f1_score = f1_score(results['y_real'], results['y_pred'])
cls_f1_score
````

## Validação Cruzada

Se você por alguma razão precisou resodar esse mesmo notebook uma segunda ou terceira vez antes chegar até aqui, pode ter percebido que os valores das métricas variam a cada nova rodada. Isso ocorre basicamente porque quando fazemos os splits de treino e teste, reservamos sempre amostras do mesmo tamanho, mas nunca os mesmos valores e, aprendendo com dados diferentes, os algoritmos têm capacidades diferentes de predição. O que fazemos então quando aguém perguntar sobre alguma de nossas métricas já que elas sempre mudam? Podemos fazer uma média de várias previsões feitas depois de treinar em dados distintos. Uma das formas mais conhecidas de fazer algo assim é a **validação cruzada em k folds**.

Basicamente dividimos o conjunto de dados em k pedaços do mesmo tamanho separamos k-1 deles para treino e que resta para validação, e trocamos vamos revezando o conjunto usado para teste até que tenhamos realizados as k rodadas de treino e teste em dados distintos.




image.png

Para saber um pouco mais sobre como funciona essa técnica, você pode dar uma lida [nesse artigo](https://drigols.medium.com/introdu%C3%A7%C3%A3o-a-valida%C3%A7%C3%A3o-cruzada-k-fold-2a6bced32a90) ou [nesse outro](https://pt.wikipedia.org/wiki/Valida%C3%A7%C3%A3o_cruzada).

Aqui no nosso notebook, vamos seguir com o [sklearn](https://scikit-learn.org/stable/) usando agora o [cross_val_score](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) com 10 amostras aleatórias.

````python
from sklearn.model_selection import cross_val_score, KFold

# tomando novamente o caso da classificação e olhando especificamente para o f1
# dividindo os dados em treino e teste
x = boston_housing_cls[['LotArea', 'SalePrice']]
y = boston_housing_cls['BinnaryOverallQual']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7)

cls = KNeighborsClassifier(n_neighbors=5) # instanciando o modelo
folds = KFold(10, shuffle=True)

results = cross_val_score(cls, x, y, scoring="f1", cv=folds)
results
````

Note que os 10 valores têm uma diferença significativa entre o máximo e mínimo. Isso indica que os dados não aparentam ser muito uniformes ja que o mesmo modelo aplicado a pedaços diferentes do mesmo dataset gera resultados tão distintos.

Em outros momentos do curso vamos aprender sobre formas de tratar esses dados pra reduzir o máximo possível essa variância que num mundo ideal deveria ser o mais próxima de zero possível.

````python
# no momento vamos nos contentar dizendo que uma maneira mais formal de reportar
# os resultados do nosso knn padrão, é dizer que ele acerta a média e associar algum intervalo 
# de confiança a esse valor

print(f"{results.mean()} ± {2*results.std()}")
````

### Encoding

````python
boston_housing_full = novo_dataset_sem_nulos.copy() # instanciando um novo dataframe ja sem nulos
boston_housing_full['BinnaryOverallQual'] = (boston_housing_full['OverallQual'].
                                             apply(lambda x: 0 if x < 8 else 1)) # reaplicando a criação de BinnaryOverallQual

boston_housing_full
````

Olhando para o dataframe inteiro, vemos que muitas das informações são textos, o que para o nosso modelo não é interessante. Na verdade nosso modelo não sabe lidar com nada que não dados numpericos. Sendo assim, a primeira coisa para ajudar o nosso modelo a performar melhor, é fazer com que ele entenda os dados.

Para tal, temos algumas possibilidades dependendo do caso. Algumas delas:
1. **Label encoding**: provavelmente a mais genérica e, consequentemente, a que geralmente escolhemos para fazer um modelos o mais simples possível, simplesmente substitui a representação textual de uma variável por um número aleatório. Pode ser o caso, por exemplo de substituir os nomes das cidades em um dataframe.

2. **One hot encoding**: podemos pensar nesse método quando temos algum tipo de categoria, com poucos valores unicos de preferência, que não aparentam ter relaçãoes lógicas entre si por exemplo uma coluna que armazena o período do dia como manhã, tarde e noite, quando transformada pelo one hot, cria 3 colunas (manhã, tarde e noite) onde um registro de manhã, passa a ser reprensentado pro 1 na coluna manhã e 0 nas outras duas, bem como noite é representada por 0 em manhã e tarde e 1 em noite.\
Daí podemos entender o porque de tomar cuidado com o número de valor únicos. Para uma variável com n valores únicos teremos na melhor das hipóteses n-1 novas colunas no dataframe.

3. **Ordinal encoding**: quando podemos ver uma relação direta entre as categorias por exemplo um modelo de notas que vai de E até A que pode ser substituido por um ranking indo de 1 a 5.

4. **Target encoding**: Nesse caso, olhamos para os valores de resposta que nós temos nos dados de treino para criar algum tipo de relação, então só é possível usar label encoder em aprendizado supervisionado (no contexto desse notebook, classificação e regressão, então podemos usar tranquilamente por aqui). Podemos por exemplo, e acho que até vamos, substituir o nome do bairro onde uma casa está localizada pela média dos preços de uma casa do mesmo nível de qualidade no mesmo bairro.
