Como sÃ£o muitas colunas, podemos tentar tirar uma primeira impressÃ£o sobre a informaÃ§Ã£o de forma mais resumida.

````python
# removendo valores nulos de forma genÃ©rica
dataset_sem_nulos = dataset_com_nulos.dropna() # joga fora todas as linas com pelo menos um valor nulo
print(f"tamanho do dataset com nulos: {dataset_com_nulos.shape}")
print(f"tamanho do dataset sem nulos: {dataset_sem_nulos.shape}")

dataset_sem_nulos.isnull().sum()[dataset_sem_nulos.isnull().sum() > 0] # sÃ©rie vazia = nenhuma coluna com nulos
# o problema Ã© que nesse caso nÃ£o sobrou linha nenhuma (0 (linhas), 81 (colunas))
````

````python
# preenchendo valores nulos com alguma mÃ©trica da propria coluna
# uma estrategia mais interessante que simplesmente jogar a linha fora por ter um nulo, 
# poderia ser, por exemplo, preencher o nulo com o valor que mais se repete naquela coluna, a moda
novo_dataset_sem_nulos = dataset_com_nulos.copy() # iniciando o dataframe a ser preeenchido
for column in dataset_sem_nulos.columns:
    novo_dataset_sem_nulos[column] = novo_dataset_sem_nulos[column].fillna(dataset_com_nulos[column].mode()[0])

print(f"tamanho do dataset sem nulos: {novo_dataset_sem_nulos.shape} \n")
novo_dataset_sem_nulos.isnull().sum()[novo_dataset_sem_nulos.isnull().sum() > 0] 

# agora sim um dataset sem nulos, mas com o tamanho do original
display(novo_dataset_sem_nulos)
````

Do mesmo modo que lÃ¡ em cima usamos as list comprehensions, aqui temos uma outra novidade: as funÃ§Ãµes lambda ou funÃ§Ãµes anÃ´nimas.

Basicamente essa Ã© uma sintaxe que resume uma funÃ§Ã£o de um argumento em uma linha:

```python
def x_na_quinta_padrao(x):
    return x ** 5
```

pode ser resumida como:

```python
lambda x: x ** 5
```

quando dentro de um contexto como no caso do apply que passa os valores dinamicamente para dentro do argumento x ou de forma bem menos usual como:

```python
lambda_x_na_quinta = lambda x: x ** 5
```

que pode ser acessada em qualquer lugar do cÃ³sigo, mas perde a sua vantagem que Ã© ser definida e invocada em um sÃ³ comando.

Nesse caso, `x_na_quinta_padrao` e `lambda_x_na_quinta` tÃªm exatamente o mesmo comportamento.

Para entender um pouco sobre essa forma de escrever funÃ§Ãµes de uma linha, vocÃª pode dar uma lida [nesse post](https://giovannireisnunes.wordpress.com/2018/08/03/funcoes-anonimas-em-python/) ou [nesse vÃ­deo](https://www.youtube.com/watch?v=EQRsJ2bKvUU) do YouTube.

## RegressÃ£o

Vamos seguir usando o dataset boston housing, mas para simplificar um pouco a anÃ¡lise, vamos selecionar apenas 2 colunas: LotArea e OverallQual para prever SalePrice.

Usaremos o modelo de [regressÃ£o linear do Sciki-learn](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

````python
boston_housing_reg = boston_housing_df[['LotArea', 'OverallQual', 'SalePrice']]
boston_housing_reg
````

### Linear

````python
from sklearn.linear_model import LinearRegression

# separando as variÃ¡veis de entrada das variÃ¡vel a ser prevista
x = boston_housing_reg[['LotArea', 'OverallQual']]
y = boston_housing_reg['SalePrice']

reg = LinearRegression() # instanciando o modelo
reg.fit(x, y) # treinando o modelo

prev_indice_1458 = reg.predict([[9717,5]])[0] # fazendo uma previsÃ£o da casa de indice 1458 do dataframe
print(prev_indice_1458) # selecionamos o indice 0 duas vezes por causa do formato de saÃ­da dos dados
````

Mas porque o modelo nÃ£o acerta uma casa que ele conhece?\
Lembrando dos vÃ­deos, uma regressÃ£o Ã© um modelo que ajusta uma curva aos dados tentando minimizar o erro entÃ£o provavelmente nÃ£o vai passar exatamente em cima de algum dos pontos usados para o treino.

````python
boston_housing_reg.loc[1458]
````

### Medidas de desempenho da regressÃ£o

Para avaliar o modelo de uma forma um pouco mais cuidadosa que apensa verificar se o indice 1458 estÃ¡ suficientemente prÃ³ximo do valor real, vamos retreinar o modelo com um conjunto de dados e testar como esse modelo se comporta lidando com dados que ele nunca viu:

````python
from sklearn.model_selection import train_test_split

x = boston_housing_reg[['LotArea', 'OverallQual']]
y = boston_housing_reg['SalePrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7) 

print(x_train.shape) # duas colunas com 70% dos dados
print(y_train.shape) # uma coluna com 70% dos dados
print(x_test.shape)
print(y_test.shape)
````

````python
reg = LinearRegression() # instanciando o modelo
reg.fit(x_train, y_train) # treinando o modelo

y_pred = reg.predict(x_test) # prevendo todos os valores de teste de uma vez

results = x_test.copy()
results['y_real'] = y_test
results['y_pred'] = y_pred

display(results.head(20)) # comparando o resultado de 20 previsoes
````

## ValidaÃ§Ã£o Cruzada

Se vocÃª por alguma razÃ£o precisou resodar esse mesmo notebook uma segunda ou terceira vez antes chegar atÃ© aqui, pode ter percebido que os valores das mÃ©tricas variam a cada nova rodada. Isso ocorre basicamente porque quando fazemos os splits de treino e teste, reservamos sempre amostras do mesmo tamanho, mas nunca os mesmos valores e, aprendendo com dados diferentes, os algoritmos tÃªm capacidades diferentes de prediÃ§Ã£o. O que fazemos entÃ£o quando aguÃ©m perguntar sobre alguma de nossas mÃ©tricas jÃ¡ que elas sempre mudam? Podemos fazer uma mÃ©dia de vÃ¡rias previsÃµes feitas depois de treinar em dados distintos. Uma das formas mais conhecidas de fazer algo assim Ã© a **validaÃ§Ã£o cruzada em k folds**.

Basicamente dividimos o conjunto de dados em k pedaÃ§os do mesmo tamanho separamos k-1 deles para treino e que resta para validaÃ§Ã£o, e trocamos vamos revezando o conjunto usado para teste atÃ© que tenhamos realizados as k rodadas de treino e teste em dados distintos.




image.png

Para saber um pouco mais sobre como funciona essa tÃ©cnica, vocÃª pode dar uma lida [nesse artigo](https://drigols.medium.com/introdu%C3%A7%C3%A3o-a-valida%C3%A7%C3%A3o-cruzada-k-fold-2a6bced32a90) ou [nesse outro](https://pt.wikipedia.org/wiki/Valida%C3%A7%C3%A3o_cruzada).

Aqui no nosso notebook, vamos seguir com o [sklearn](https://scikit-learn.org/stable/) usando agora o [cross_val_score](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) com 10 amostras aleatÃ³rias.

````python
from sklearn.model_selection import cross_val_score, KFold

# tomando novamente o caso da classificaÃ§Ã£o e olhando especificamente para o f1
# dividindo os dados em treino e teste
x = boston_housing_cls[['LotArea', 'SalePrice']]
y = boston_housing_cls['BinnaryOverallQual']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7)

cls = KNeighborsClassifier(n_neighbors=5) # instanciando o modelo
folds = KFold(10, shuffle=True)

results = cross_val_score(cls, x, y, scoring="f1", cv=folds)
results
````

Note que os 10 valores tÃªm uma diferenÃ§a significativa entre o mÃ¡ximo e mÃ­nimo. Isso indica que os dados nÃ£o aparentam ser muito uniformes ja que o mesmo modelo aplicado a pedaÃ§os diferentes do mesmo dataset gera resultados tÃ£o distintos.

Em outros momentos do curso vamos aprender sobre formas de tratar esses dados pra reduzir o mÃ¡ximo possÃ­vel essa variÃ¢ncia que num mundo ideal deveria ser o mais prÃ³xima de zero possÃ­vel.

````python
# no momento vamos nos contentar dizendo que uma maneira mais formal de reportar
# os resultados do nosso knn padrÃ£o, Ã© dizer que ele acerta a mÃ©dia e associar algum intervalo 
# de confianÃ§a a esse valor

print(f"{results.mean()} Â± {2*results.std()}")
````

Olhando para o dataframe inteiro, vemos que muitas das informaÃ§Ãµes sÃ£o textos, o que para o nosso modelo nÃ£o Ã© interessante. Para tal, temos algumas possibilidades dependendo do caso. Algumas delas:
1. **Label encoding**: provavelmente a mais genÃ©rica e, consequentemente, a que geralmente escolhemos para fazer um modelos o mais simples possÃ­vel, simplesmente substitui a representaÃ§Ã£o textual de uma variÃ¡vel por um nÃºmero aleatÃ³rio. 2. **One hot encoding**: podemos pensar nesse mÃ©todo quando temos algum tipo de categoria, com poucos valores unicos de preferÃªncia, que nÃ£o aparentam ter relaÃ§Ã£oes lÃ³gicas entre si por exemplo uma coluna que armazena o perÃ­odo do dia como manhÃ£, tarde e noite, quando transformada pelo one hot, cria 3 colunas (manhÃ£, tarde e noite) onde um registro de manhÃ£, passa a ser reprensentado pro 1 na coluna manhÃ£ e 0 nas outras duas, bem como noite Ã© representada por 0 em manhÃ£ e tarde e 1 em noite.\
DaÃ­ podemos entender o porque de tomar cuidado com o nÃºmero de valor Ãºnicos. Para uma variÃ¡vel com n valores Ãºnicos teremos na melhor das hipÃ³teses n-1 novas colunas no dataframe.

3. **Ordinal encoding**: quando podemos ver uma relaÃ§Ã£o direta entre as categorias por exemplo um modelo de notas que vai de E atÃ© A que pode ser substituido por um ranking indo de 1 a 5.

4. **Target encoding**: Nesse caso, olhamos para os valores de resposta que nÃ³s temos nos dados de treino para criar algum tipo de relaÃ§Ã£o, entÃ£o sÃ³ Ã© possÃ­vel usar label encoder em aprendizado supervisionado (no contexto desse notebook, classificaÃ§Ã£o e regressÃ£o, entÃ£o podemos usar tranquilamente por aqui). Podemos por exemplo, e acho que atÃ© vamos, substituir o nome do bairro onde uma casa estÃ¡ localizada pela mÃ©dia dos preÃ§os de uma casa do mesmo nÃ­vel de qualidade no mesmo bairro.
